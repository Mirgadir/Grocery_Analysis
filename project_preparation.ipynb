```python
grocery_analysis_data = grocery_data[['Category', 'Sub_category', 'Product_Group',
'Product_Name', 'Package_price', 'Price_per_unit', 'package_size',
'in_stock', 'Retail_price', 'Brand', 'RunDate', 'unit_price', 'unit_price_unit', 'state', 'city']]
```

```python
grocery_analysis_data['Price_Markup'] =  (grocery_analysis_data['Retail_price']/ grocery_analysis_data['Package_price'] * 100) - 100
```
